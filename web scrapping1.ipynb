```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re 

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']

base_url = 'https://books.toscrape.com/catalogue/'
li1 = []
for i in allPages:
    response = requests.get(i)
    data = BeautifulSoup(response.text, 'html.parser')
    li1.append(data)
# li1
li2 = []
for j in li1:
    li2.append(j.find_all('h3'))
li2

li4 = []
for j in li2:
    for i in j:
        final_url = base_url+i.a['href']
        response = requests.get(final_url)
        data = BeautifulSoup(response.text, 'html.parser')
        title = data.h1.string
        price = data.find(class_='price_color').string
        qty = data.find(class_='instock availability').contents[-1].strip()
        li4.append([title, final_url, price, qty])

for i in li4:
    i[2] = float(re.search('[\d.]+', i[2]).group())
    i[3] = int(re.search('\d+', i[3]).group())

df = pd.DataFrame(li4, columns=['Title', 'Link', 'Price', 'Quantity in Stock'])
df['Title'], df['Link'], df['Price'], df['Quantity in Stock']
```
